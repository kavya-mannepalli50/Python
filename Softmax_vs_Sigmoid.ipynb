```python
import numpy as np
import matplotlib.pyplot as plt

logits = np.random.randn(5)
```

```python
softmax = np.exp(logits) / np.sum(np.exp(logits))
```

```python
plt.bar(range(len(softmax)), softmax)
plt.xlabel("Class")
plt.ylabel("Probability")
plt.title("Softmax Probability Distribution")
plt.show()
```

```python
vocab = ["I", "like", "machine", "translation"]
```

```python
logits = np.array([1.2, 2.5, 0.3, 1.0])
```

```python
probs = np.exp(logits) / np.sum(np.exp(logits))
list(zip(vocab, probs))
```

```python
next_word = vocab[np.argmax(probs)]
next_word
```

```python
next_word = np.random.choice(vocab, p=probs)
next_word
```
